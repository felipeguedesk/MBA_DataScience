## Exercício 6) 

Analise a base de dados "Exercicio" novamente. Agora vamos buscar por problemas dos tipos errôneos e outliers no atributo "Years of Experience":


```python
data_m = pd.read_csv("./dados/TACTD-01-Exercicio.csv")

print(data_m['Years of Experience'].describe(), '\n')

print("Listando os valores não nulos e fora do intervalo esperado e sua quantidade")
print("Valor\t Quantidade")
# exibindo valores nulos/nan
print("nan\t ", np.sum(data_m['Years of Experience'].isnull()))
#exibindo valores nao nulos e num intervalo improvável
for value in data_m['Years of Experience'].unique():
    if (value == value and (value < 0 or value > 30)):
        print(value, end='\t  ')
        print(np.sum(data_m['Years of Experience'] == value))
        
plt.figure(figsize=(12,4))
plt.subplot(121); data_m.boxplot('Years of Experience')
plt.subplot(122); data_m['Years of Experience'].hist()
```

```python
data_m[data_m['Years of Experience'].isnull()]
```

```python
data_m[(data_m['Years of Experience'] < 0) | (data_m['Years of Experience'] > 30)]
```
