### Exercício 5)
A página ['http://dados.prefeitura.sp.gov.br/pt_PT/dataset'](http://dados.prefeitura.sp.gov.br/pt_PT/dataset) do portal possui uma barra de busca para que se realize consultas às bases existentes. O método <font color='blue'>get</font> pode realizar uma requisição onde parâmetros são enviados para a barra de busca.
Por exemplo:
```python
rq.get('http://dados.prefeitura.sp.gov.br/pt_PT/dataset',
       params={'q': 'seguranca'})
```
envia a palavra 'seguranca' para a barra de busca, retornando um conteúdo HTML com o resultado da busca.

Faça uma requisição enviando como parâmetro de busca a palavra "saude". Faça um parsing no conteúdo retornado e extraia os nomes e as URLs dos repositórios retornados. 
Por exemplo, 

```html
Cadastro dos Estabelecimentos de Saúde

Este conjunto de dados contém informações sobre todos os equipamentos municipais de saúde, fruto de um esforço realizado pelo Departamento de Produção e Análise de Informação...
```
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="background-color:green">XLS</span>
<span style="background-color:orange">CSV</span>

corresponde ao primeiro repositório listado no conteúdo retornado. 
O nome do repositório é "Cadastro dos Estabelecimentos de Saúde" o qual está linkado à página [/pt_PT/dataset/cadastro-dos-estabelecimentos-de-saude](http://dados.prefeitura.sp.gov.br/pt_PT/dataset/cadastro-dos-estabelecimentos-de-saude).
Armaze o conteúdo extraído em um dicionário onde a chave é o nome do repositório e o valor é a URL correspondente.


````python
import requests as rq
from bs4 import BeautifulSoup

# podemos fazer uma lista de queries
for query in ['seguranca', 'saude']:
    response = rq.get('http://dados.prefeitura.sp.gov.br/pt_PT/dataset', params={'q': query})
    print(f"Status Code: {response.status_code}")
    soup = BeautifulSoup(response.text, 'html.parser')
    all_repos = soup.find_all('h3', {'class': 'dataset-heading'})
    repos_dict = {}
    for repo in all_repos:
        link = repo.find('a')
        title = link.text
        url = link['href']
        repos_dict[title] = url
    print(f"Query: {query}")
    print(f"Repos: {repos_dict}")
    print()
````

### Exercício 6)
Cada uma das URLs obtidas no exercício 5) leva a um repositório contendo arquivos para download. Escreva um código para realizar uma requisição em todas as páginas apontadas pelas URLs obtidas no exercício 5). Armazene o resultado das requisições em uma lista chamada `repos`, ou seja, cada ítem da lista corresponde ao conteúdo retornado pela requisição feita a um repositório.

````python
base_url = "http://dados.prefeitura.sp.gov.br"
repos_html = {}
for name, url in repos_dict.items():
    response = rq.get(f"{base_url}{url}")
    if response.status_code == 200:
        repos_html[name] = response
repos_html
````

### Exercício 7)
Os conteúdos armazenados na lista `repos` são páginas HTML. Realize um parsing em cada um dos ítens da lista `repos` buscando pelo nome do repositório, a data de criação e a data da última atualização do repositório. Imprima tais informações.

````python
def filter_text(text):
    return 'Última Atualização' in text

for name, response in repos_html.items():
    soup = BeautifulSoup(response.text, 'html.parser')
    module_content = soup.find('article', {'class': 'module'}).find('div', {'class': 'module-content'})
    title = module_content.find('h1').text
    print(f"Title: {title.strip()}")
    add_info = soup.find('section', {'class': 'additional-info'}).find_all('th')
    
    for header in add_info:
        if 'Última Atualização' in header.text:
            update_date = header.find_next().text
            print(f"Última Atualização: {update_date}")
        if 'Criado' in header.text:
            update_date = header.find_next().text
            print(f"Data Criação: {update_date}")
            
    update_header = soup.find('section', class_='additional-info').find('th', text=filter_text)
    update_date = update_header.find_next().text
    print(f"Última Atualização: {update_date}")
    update_header = soup.find('section', class_='additional-info').find('th', text='Criado')
    update_date = update_header.find_next().text
    print(f"Data Criação: {update_date}")
    
    print()
````

### Exercício 8)
Para cada repositório da lista `repos` imprima a quantidade total de arquivos XLS, CSV, etc. disponíveis para download. 

````python
for name, response in repos_html.items():
    soup = BeautifulSoup(response.text, 'html.parser')
    resources = soup.find('section', id='dataset-resources').find_all('li', class_='resource-item')
    print(f"{name}: {len(resources)}")
#     print(resources)
````

### Exercício 9)
Alguns grupos de dados possuem muitos repositórios. Escreva um código para imprimir os nomes de todos os repositórios do grupo "Orçamento e Gestão". 

**DICA**: A URL para esse grupo pode ser encontrada no Exercício 4.

````python
page_counter = 1
while True:
    response = rq.get('http://dados.prefeitura.sp.gov.br/group/financas', params={'page': page_counter})
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        list_items = soup.find_all('li', class_='dataset-item')
        if not list_items:
            break
        for list_item in list_items:
            heading = list_item.find('h3', class_='dataset-heading').find('a').text
            print(heading)
        print()
        page_counter += 1
````
