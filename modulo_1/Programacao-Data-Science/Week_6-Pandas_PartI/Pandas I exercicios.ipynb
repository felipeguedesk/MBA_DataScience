### Exercício 3)

### Exercício 8)
Construa um dicionário onde as chaves são os países que já participaram de uma final e os valores são o número de vezes que o país participou da final.

```python
# SOLUÇÃO

teams_in_final = np.unique(list(worldCupTable['Winner'].unique())+list(worldCupTable['RunnersUp'].unique()))
print(teams_in_final)
finalist_dict = {country:(len(worldCupTable.query(f"Winner == '{country}'")) + len(worldCupTable.query(f"RunnersUp == '{country}'"))) for country in teams_in_final}
print('Países que participaram de uma final:\n')       
for c,v in finalist_dict.items():
  print(c,': ',v)
```

### Exercício 9) 
Construa um dicionário com todos os países que já participaram de uma semi-final de copa (os quatro finalistas), onde a chave é o país e o valor corresponde ao número de vezes que o país foi um semi-finalista.

```python
# SOLUÇÃO
campeoes = worldCupTable['Winner'].value_counts()
vices = worldCupTable['RunnersUp'].value_counts()
terceiros = worldCupTable['Third'].value_counts()
quartos = worldCupTable['Fourth'].value_counts()
uniqueSemiFinalists = np.unique(np.hstack((campeoes.index,vices.index, terceiros.index, quartos.index)))
finalist_dict = {unique:0 for unique in uniqueSemiFinalists}
finalists = [campeoes,vices, terceiros, quartos]
def getCount(country):
    countryCount = 0
    for finalist in finalists:
        if (country in finalist):
            countryCount += finalist[country]
    return countryCount
for country in uniqueSemiFinalists:
    finalist_dict[country] = getCount(country)
print(finalist_dict)

```
